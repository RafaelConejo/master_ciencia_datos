PPCA responde al siguiente modelo probabil칤stico (tomamos los datos normalizados):

\begin{align*}
p(\mathbf{z}) &= \mathcal{N}(\mathbf{z} \mid \mathbf{0}, \mathbf{I})\\
p(\mathbf{x} \mid \mathbf{z}) &= \mathcal{N}(\mathbf{x} \mid \mathbf{W} \mathbf{z},\sigma^2 \mathbf{I})\
\end{align*}

La intuici칩n detr치s de este modelo es como sigue:

1. Para cada observaci칩n $\mathbf{x}$ tenemos una variable latente 'impl칤cita' (que no observamos) $\mathbf{z}$ con $M$ componentes ($M$ es el n칰mero de componentes principales escogido). Introducimos una distribuci칩n a priori (gaussiana) $p(\mathbf{z})$ sobre la variable latente. De ella extraemos una muestra $\mathbf{z}$.
2. Con esa muestra generamos nuestras observaciones usando el modelo de observaci칩n $p(\mathbf{x} \mid \mathbf{z})$ (en nuestro caso lineal y con ruido Gaussiano).

Para que te situes: en la clase de teor칤a sobre PPCA hemos supuesto que tenemos $N$ ejemplos, cada uno con $D$ componentes. Usaremos $M$ para notar el n칰mero de componentes principales. Observa que $\mathbf{W}$ tiene tama침o $D \times M$ y $\sigma^2$ controla la varianza de la distribuci칩n condicionada (cu치nto ruido tenemos en las observaciones).

## C치lculo de los estimadores de m치xima verosimilitud en PPCA

Calculados los autovectores y autovalores de la matriz de covarianza muestral, podemos calcular la matriz $\mathbf{W}$ cuyas columnas generan el subespacio principal. Como sabemos, su estimador de m치xima verosimilitud (max-likelihood) es:

\begin{equation*}
\mathbf{W}_{\mathrm{ML}} = \mathbf{U}(\mathbf{L}-\sigma^2 \mathbf{I})^{1/2} \mathbf{R} \\
\end{equation*}

donde $\mathbf{U}$ es una matriz $D \times M$ cuyas columnas corresponden a los $M$ autovectores o autocaras asociadas a los $M$ mayores autovalores de la matriz de covarianza muestral, $\mathbf{L}$ es una matriz diagonal que contiene los autovalores correspondientes y $\mathbf{R}$ es una matriz ortonormal de dimensiones $M \times M$ (una matriz de rotaci칩n arbitraria). Por simplicidad, supondremos que $\mathbf{R}$ es la matriz identidad.

Como estimador de $\sigma^2$ usaremos su estimador de m치xima verosimilitud que como sabes es

\begin{equation*}
\sigma_{\mathrm{ML}}^2 = \frac{1}{D-M}\sum_{i=M+1}^D\lambda_i \\
\end{equation*}

# Detecci칩n de anomal칤as mediante $ \log p(\mathbf{x})$

\begin{equation*}

\end{equation*} 

Esta cantidad se llama en ingl칠s *log likelihood* y mide la verosimilitud de una muestra. Observa que $\log p(\mathbf{x} \mid \mathbf{W}_{\mathrm{ML}}, \sigma_{\mathrm{ML}}^2)$ es menor conforme menos probable sea la muestra $\mathbf{x}$.
