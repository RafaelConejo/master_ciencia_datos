* Create a function to convert ºF to ºC and in (inches) to mm.

```r
convert_F_to_C <- function(x){
    x <- (x-32) / 1.8
    x
}
convert_in_to_mm <- function(x){
    x <- as.numeric(x)
    x[is.na(x)] <- 0
    
    x <- x * 25.4
    x
}


str(data)
data_C <- data

data_C[,2:8]<- apply(data[,2:8], 2, convert_F_to_C)

#Da error porque se trata dentro del apply con los valores NA, intenta hacerlo fuera
data_C[,9:12]<- apply(data[,9:12], 2, convert_in_to_mm)
data_C
```
