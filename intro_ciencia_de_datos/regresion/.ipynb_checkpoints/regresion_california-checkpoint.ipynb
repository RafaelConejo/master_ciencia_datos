# Regresi√≥n con datos dataset California

```r
# Comienzo con la lectura de los datos
california = read.csv("california.dat", comment.char="@", header=FALSE)

names(california) <- c("Longitude", "Latitude", "HousingMedianAge",
                 "TotalRooms", "TotalBedrooms", "Population", "Households",
                 "MedianIncome", "MedianHouseValue")

#Asignacion automatica, facilita el acceso a los campos
n <- length(names(california)) - 1
names(california)[1:n] <- paste ("X", 1:n, sep="")
names(california)[n+1] <- "Y"
```

Se observan las relaciones de todas las variables entre si y respecto a la salida (Y).
Observamos cual se ajusta mejor a un modelo lineal

```r
pairs(california, upper.panel=NULL)
```

Se observa posible correlaci√≥n entre algunas de las variables.
Respecto a la salida la variable que parece tener un comportamiento m√°s lineal es la X8 (MedianIncome), por ello a continuaci√≥n se probar√° a crear un modelo lineal con dicha variable.

```r
fit1 = lm(Y ~ X8, data=california)
summary(fit1)
plot(Y ~ X8, data=california)
abline(fit1,col="red")
```

Efectivamente es una variable importante para la predicci√≥n del modelo, pero el valor del R ajustado obtenido es bajo, por ello se probar√° a continuaci√≥n a√±adiendo m√°s variables.

Se crear√° para ello un primer modelo con todas las variables y se ir√°n retirando las variables que se cosideren que peor predicen (el peor p-value)

```r
fit2 = lm(Y ~ ., data=california)
summary(fit2)
```

Se observa la importancia de todas las variables para predecir el modelo.

Se prueba a eliminar la variable con mayor p-value (aunque sigue siendo un valor muy peque√±o)

```r
fit3 = lm(Y ~ . -X7, data=california)
summary(fit3)
```

Se observa que el modelo empeora levemente.

Es posible tambi√©n realizar combinaciones lineales m√°s complejas

```r
fit4 <- lm(Y ~ X8 * X5, data=california)
summary(fit4)
```

```r
fit5 <- lm(Y ~ X8 + I(X3 * X5), data=california)
summary(fit5)
```

Es posible graficar un modelo ajustado para observar su comportamiento

```r
fit6 = lm(Y ~ X8, california)
summary(fit6)
plot(california$Y ~ california$X8, pch='.')
points(california$X8, fitted(fit6), col="red",pch=20)
```

Partiendo de un modelo ajustado, este puede ser utilizado para la predicci√≥n de nuevos datos

```r
y_pred <- predict(fit6, california)

#Realizada la predici√≥n es posible calcular la ra√≠z del ECM
sqrt(sum(abs(california$Y - y_pred)^2) / length(y_pred))
```

## Parte 2

```r
install.packages('kknn')
```

```r
library('kknn')
fitknn1 <- kknn(Y ~ ., california, california)
```

Obtenemos informaci√≥n sobre el objeto obtenido

```r
names(fitknn1)
```

Al igual que antes es posible visualizar el resultado obtenido

```r
plot(california$Y ~ california$X8)
points(california$X8, fitknn1$fitted.values, col="blue", pch=20)
```

y tambien es posible el calculo manueal de la raiz del ECM y predicciones

```r
yprime = fitknn1$fitted.values

sqrt(sum((california$Y - yprime) ^ 2) / length(yprime))
```

Tambi√©n es posible realizar interacciones entre variables en knn

```r
fitknn2 <- kknn(Y ~ X8 + I(X5 * X7), california, california)
yprime <- fitknn2$fitted.values
sqrt(sum((california$Y - yprime) ^ 2) / length(yprime)) #RMSE
```

Aplicaci√≥n de k-fold cross-validation LM para todas las variables

```r
nombre <- "california"
run_lm_fold <- function(i, x, tt = "test") {
    file <- paste(x, "-5-", i, "tra.dat", sep="")
    x_tra <- read.csv(file, comment.char="@", header=FALSE)
    file <- paste(x, "-5-", i, "tst.dat", sep="")
    x_tst <- read.csv(file, comment.char="@", header=FALSE)
    In <- length(names(x_tra)) - 1
    names(x_tra)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tra)[In+1] <- "Y"
    names(x_tst)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tst)[In+1] <- "Y"
    if (tt == "train") {
        test <- x_tra
    }
    else {
        test <- x_tst
    }
    fitMulti=lm(Y~.,x_tra)
    yprime=predict(fitMulti,test)
    sum(abs(test$Y-yprime)^2)/length(yprime) ##MSE
}
lmMSEtrain<-mean(sapply(1:5,run_lm_fold,nombre,"train"))
lmMSEtest<-mean(sapply(1:5,run_lm_fold,nombre,"test"))
```

Aplicaci√≥n de k-fold cross-validation KNN para todas las variables

```r
nombre <- "california"
run_knn_fold <- function(i, x, tt = "test") {
    file <- paste(x, "-5-", i, "tra.dat", sep="")
    x_tra <- read.csv(file, comment.char="@", header=FALSE)
    file <- paste(x, "-5-", i, "tst.dat", sep="")
    x_tst <- read.csv(file, comment.char="@", header=FALSE)
    In <- length(names(x_tra)) - 1
    names(x_tra)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tra)[In+1] <- "Y"
    names(x_tst)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tst)[In+1] <- "Y"
    if (tt == "train") {
        test <- x_tra
    }
    else {
        test <- x_tst
    }
    fitMulti=kknn(Y~.,x_tra,test)
    yprime=fitMulti$fitted.values
    sum(abs(test$Y-yprime)^2)/length(yprime) ##MSE
}
knnMSEtrain<-mean(sapply(1:5,run_knn_fold,nombre,"train"))
knnMSEtest<-mean(sapply(1:5,run_knn_fold,nombre,"test"))
```

Al comparar los resultados se observa como los obtenidos con KNN son mejores que los obtenidos con LM, puesto que tanto en el train como test el valor de MSE es menor con KNN.

```r
lmMSEtrain
knnMSEtrain

lmMSEtest
knnMSEtest
```

Repitamos esta comparativa pero con diferentes interacciones entre variables y no con todas las variables como en el caso anterior

```r
nombre <- "california"
run_lm_fold <- function(i, x, tt = "test") {
    file <- paste(x, "-5-", i, "tra.dat", sep="")
    x_tra <- read.csv(file, comment.char="@", header=FALSE)
    file <- paste(x, "-5-", i, "tst.dat", sep="")
    x_tst <- read.csv(file, comment.char="@", header=FALSE)
    In <- length(names(x_tra)) - 1
    names(x_tra)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tra)[In+1] <- "Y"
    names(x_tst)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tst)[In+1] <- "Y"
    if (tt == "train") {
        test <- x_tra
    }
    else {
        test <- x_tst
    }
    fitMulti=lm(Y~. +I(X1^2)+I(X6^2)+I(X8^2)+I(X8^3)
                    +I(X8^4)+X7*X8*X4*X5*X6, x_tra)
    yprime=predict(fitMulti,test)
    sum(abs(test$Y-yprime)^2)/length(yprime) ##MSE
}
lmMSEtrain<-mean(sapply(1:5,run_lm_fold,nombre,"train"))
lmMSEtest<-mean(sapply(1:5,run_lm_fold,nombre,"test"))

nombre <- "california"
run_knn_fold <- function(i, x, tt = "test") {
    file <- paste(x, "-5-", i, "tra.dat", sep="")
    x_tra <- read.csv(file, comment.char="@", header=FALSE)
    file <- paste(x, "-5-", i, "tst.dat", sep="")
    x_tst <- read.csv(file, comment.char="@", header=FALSE)
    In <- length(names(x_tra)) - 1
    names(x_tra)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tra)[In+1] <- "Y"
    names(x_tst)[1:In] <- paste ("X", 1:In, sep="")
    names(x_tst)[In+1] <- "Y"
    if (tt == "train") {
        test <- x_tra
    }
    else {
        test <- x_tst
    }
    fitMulti=kknn(Y~. +I(X1^2)+I(X6^2)+I(X8^2)+I(X8^3)
                    +I(X8^4)+X7*X8*X4*X5*X6, x_tra, test)
    yprime=fitMulti$fitted.values
    sum(abs(test$Y-yprime)^2)/length(yprime) ##MSE
}
knnMSEtrain<-mean(sapply(1:5,run_knn_fold,nombre,"train"))
knnMSEtest<-mean(sapply(1:5,run_knn_fold,nombre,"test"))


lmMSEtrain
knnMSEtrain

lmMSEtest
knnMSEtest
```

Por √∫ltimo, utilizaremos el text de wilcoxon para ver cual de los dos modelos anteriores es mejor, comparando el modelo lineal con el modelo KNN.

```r
tablatst = cbind(knnMSEtest,lmMSEtest)

##TABLA NORMALIZADA - lm (other) vs knn (ref) para WILCOXON
# + 0.1 porque wilcox R falla para valores == 0 en la tabla
difs <- (tablatst[,1] - tablatst[,2]) / tablatst[,1]
wilc_1_2 <- cbind(ifelse (difs<0, abs(difs)+0.1, 0+0.1), ifelse (difs>0, abs(difs)+0.1, 0+0.1))
colnames(wilc_1_2) <- c(colnames(tablatst)[1], colnames(tablatst)[2])
head(wilc_1_2)

#Aplicaci√≥n del test de WILCOXON
LMvsKNNtst <- wilcox.test(wilc_1_2[,1], wilc_1_2[,2], alternative = "two.sided", paired=TRUE)
Rmas <- LMvsKNNtst$statistic
pvalue <- LMvsKNNtst$p.value
LMvsKNNtst <- wilcox.test(wilc_1_2[,2], wilc_1_2[,1], alternative = "two.sided", paired=TRUE)
Rmenos <- LMvsKNNtst$statistic
Rmas
Rmenos
pvalue
```

Los resultados nos indican que si hay diferencia significativas entre ambosy que el primero de ellos es el diferente. Efectivamente el modelo con KNN es mejor que el de LM.
